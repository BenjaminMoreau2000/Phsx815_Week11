
# Demonstration of k-means assumptions

This example is meant to illustrate situations where k-means will produce
unintuitive and possibly unexpected clusters. In the first three plots, the
input data does not conform to some implicit assumption that k-means makes and
undesirable clusters are produced as a result. In the last plot, k-means
returns intuitive clusters despite unevenly sized blobs.


```python
print(__doc__)

#I took part of this code from an example on the sklearn site, and here is the original signature of the person who wrote it

# Author: Phil Roth <mr.phil.roth@gmail.com>
# License: BSD 3 clause


import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

plt.figure(figsize=(12, 12))
print("S")
n_samples = 1500
random_state = 170

#
#Manufacture some clusters artificially
#
print("Decide how many clusters you would like to make:")

NC = int(input())
Centers=[]
Uncs=[]
for i in range(NC):
  
  print("What should cluster",i," have as stdev in its mean:")
  Uncs.append(int(input()))
X, y = make_blobs(n_samples=n_samples, random_state=random_state,centers=NC,cluster_std=Uncs)#Genereate the clusters based on the inputs



# correct number of clusters
y_pred = KMeans(n_clusters=NC, random_state=random_state).fit_predict(X)

plt.scatter(X[:,0], X[:,1], c=y,label="True blobs",alpha=.7,s=150,marker="^")
plt.scatter(X[:,0], X[:,1], c=y_pred,label="Model fit",alpha=1,s=30)

plt.legend()
plt.title("Kmeans with "+str(NC)+" Blobs")

plt.show()
```
